# Tomar de un directorio

```python
import matplotlib.pyplot as plt
```

```python
import cv2
import os

# Cargar el modelo
model = YOLO(r"C:\Users\paco2\Documents\GitHub\MachineLearning\trainig\view\train2\weights\epoch39.pt")

# Directorio de las imágenes
input_dir = r"C:\Users\paco2\Documents\GitHub\MachineLearning\trainig\test_images"
output_dir = r"C:\Users\paco2\Documents\GitHub\MachineLearning\trainig\output"

# Crear el directorio de salida si no existe
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Procesar todas las imágenes en el directorio de entrada
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Cargar la imagen
        img_path = os.path.join(input_dir, filename)
        orig_img = cv2.imread(img_path)

        # Realizar la predicción
        results = model.predict(source=orig_img)

        # Procesar los resultados
        for result in results:
            # Obtener los objetos detectados
            boxes = result.boxes

            # Dibujar los rectángulos de detección en la imagen original
            for box in boxes:
                x1, y1, x2, y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                cv2.rectangle(orig_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Obtener el tamaño de la ventana
        window_width = 800
        window_height = 600

        # Redimensionar la imagen para que quepa en la ventana
        resized_img = cv2.resize(orig_img, (window_width, window_height), interpolation=cv2.INTER_AREA)

        # Guardar la imagen en el directorio de salida
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, resized_img)

        # Mostrar la imagen con los rectángulos de detección
        plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
        plt.title("Detección")
        plt.show()
```
