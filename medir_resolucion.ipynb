```python
import os
from PIL import Image
import numpy as np
def obtener_calidad_y_resolucion(imagen):
  """Obtiene la calidad y la resolución de una imagen."""
  width, height = imagen.size
  resolucion = f"{width}x{height}"
  # Puedes usar librerías de análisis de imágenes para obtener una medida de calidad,
  # pero aquí solo se considera la resolución como indicador de calidad.
  calidad = "Alta" if width >= 1920 and height >= 1080 else "Media" if width >= 1280 and height >= 720 else "Baja"
  return calidad, resolucion

def obtener_nivel_brillo(imagen):
  """Obtiene el nivel de brillo promedio de una imagen."""
  # Convertir la imagen a escala de grises
  img_gray = imagen.convert('L')
  # Calcular el brillo promedio
  brillo = np.mean(np.array(img_gray))
  return brillo

# Cargar las imágenes de la carpeta
carpeta_imagenes = '/home/SafeCross/informe/train/images'
imagenes = [Image.open(os.path.join(carpeta_imagenes, f)) for f in os.listdir(carpeta_imagenes) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Analizar las imágenes
for imagen in imagenes:
  calidad, resolucion = obtener_calidad_y_resolucion(imagen)
  brillo = obtener_nivel_brillo(imagen)
  print(f"Imagen: {imagen.filename}")
  print(f"Calidad: {calidad}")
  print(f"Resolución: {resolucion}")
  print(f"Brillo: {brillo:.2f}")
  print("----------------------")
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

def obtener_calidad_y_resolucion(imagen):
  """Obtiene la calidad y la resolución de una imagen."""
  width, height = imagen.size
  resolucion = f"{width}x{height}"
  calidad = "Alta" if width >= 1920 and height >= 1080 else "Media" if width >= 1280 and height >= 720 else "Baja"
  return calidad, resolucion

def obtener_nivel_brillo(imagen):
  """Obtiene el nivel de brillo promedio de una imagen."""
  img_gray = imagen.convert('L')
  brillo = np.mean(np.array(img_gray))
  return brillo

# Cargar las imágenes de la carpeta
carpeta_imagenes = '/home/SafeCross/informe/train/images'
imagenes = [Image.open(os.path.join(carpeta_imagenes, f)) for f in os.listdir(carpeta_imagenes) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Analizar las imágenes y contar las imágenes por clase
clases = {'Alta': 0, 'Media': 0, 'Baja': 0}
brillos = []
resoluciones = []

for imagen in imagenes:
  calidad, resolucion = obtener_calidad_y_resolucion(imagen)
  brillo = obtener_nivel_brillo(imagen)
  clases[calidad] += 1
  brillos.append(brillo)
  resoluciones.append(resolucion)

# Mostrar la cantidad de imágenes por clase
print("Cantidad de imágenes por clase:")
for clase, cantidad in clases.items():
  print(f"{clase}: {cantidad}")

# Graficar la distribución de las clases
plt.figure(figsize=(10, 5))
plt.bar(clases.keys(), clases.values())
plt.xlabel("Clase de calidad")
plt.ylabel("Cantidad de imágenes")
plt.title("Distribución de la calidad de las imágenes")
plt.show()

# Graficar la distribución del brillo
plt.figure(figsize=(10, 5))
plt.hist(brillos, bins=20)
plt.xlabel("Nivel de brillo")
plt.ylabel("Cantidad de imágenes")
plt.title("Distribución del brillo de las imágenes")
plt.show()

# Graficar la distribución de las resoluciones (solo se muestra la cantidad por resolución)
resoluciones_unicas, conteo = np.unique(resoluciones, return_counts=True)
plt.figure(figsize=(10, 5))
plt.bar(resoluciones_unicas, conteo)
plt.xlabel("Resolución")
plt.ylabel("Cantidad de imágenes")
plt.title("Distribución de las resoluciones de las imágenes")
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x
plt.show()
```

```python
# Calcular la varianza de los datos
varianza_brillo = np.var(brillos)
varianza_resoluciones = np.var(resoluciones)

print(f"Varianza del brillo: {varianza_brillo:.2f}")
print(f"Varianza de las resoluciones: {varianza_resoluciones:.2f}")
```
