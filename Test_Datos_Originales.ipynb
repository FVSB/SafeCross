```
import zipfile
import os

# Ruta del archivo ZIP en Google Drive
zip_path = '/content/drive/MyDrive/dataset.zip'

# Crea un directorio temporal para extraer los archivos
os.makedirs('/content/temp', exist_ok=True)

# Abre el archivo ZIP y extrae todos los archivos en el directorio temporal
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('/content/temp')

# Imprime los nombres de los archivos extra√≠dos
for root, dirs, files in os.walk('/content/temp'):
    for file in files:
        print(os.path.join(root, file))
```

```
import shutil
import glob
import os

# Ruta del directorio actual
current_dir = '/content/temp/crosswalk dataset'

# Ruta del directorio de destino
destination_dir = '/content/temp/crosswalk dataset/imagea'

# Crea el directorio de destino si no existe
os.makedirs(destination_dir, exist_ok=True)

# Encuentra todos los archivos JPG en el directorio actual
jpg_files = glob.glob(os.path.join(current_dir, '*.jpg'))

# Mueve todos los archivos JPG al directorio de destino
for file in jpg_files:
    shutil.move(file, destination_dir)

print(f"Se han movido {len(jpg_files)} archivos JPG a {destination_dir}")
```

```
import zipfile
import os

carpeta_a_comprimir = '/content/runs/detect/'
ruta_destino = '/content/drive/MyDrive/ML_Project/Salve_Test_Completo'
nombre_archivo_zip = 'test_completo.zip'
ruta_completa_zip = os.path.join(ruta_destino, nombre_archivo_zip)

def zipdir(path, ziph):
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file),
                       os.path.relpath(os.path.join(root, file),
                                       os.path.join(path, '..')))

# Aseg√∫rate de que la carpeta de destino exista
os.makedirs(ruta_destino, exist_ok=True)

with zipfile.ZipFile(ruta_completa_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipdir(carpeta_a_comprimir, zipf)

print(f"Archivo zip creado: {ruta_completa_zip}")
```
